```python
# re -> regular expressions
import re
import pandas as pd
```

```python
with open ("Whatsapp_Chat_with_FSDS_2022.txt", 'r', encoding='utf-8') as f:
    data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
# use regex101.com to generate regular expressions
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'date':dates})
df
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%d/%m/%Y, %H:%M - ')
df.head(10)
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split(':\s', message)
    if(entry[1:]):  # user name
        users.append(entry[0]) #creating a list and inserting all the users
        messages.append(entry[1]) #second part of the string(after colon), adding that in messages
    else:
        users.append('group_notification') #if any string doesn't contain colon, then it'll be a group notification, 
                                           #so appending that in users
        messages.append(entry[0]) #rest will remain as it is in messages
        
df['user'] = users #creating new column named user
df['message'] = messages #creating new column named message
```

```python
# remove stop words
words = []
for message in temp['message']:
    for word in message.lower().split(" "):
        if word not in stop_words:
            words.append(word)
```

```python
# top 20 words of the whatsapp group
# counter will give the frequency of each and every word used in the group
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```

```python
# At what time are the users most active in a week
period = []
for hour in df['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str("00"))
    elif hour == 0:
        period.append(str("00") + "-" + str(hour+1))
    else:
        if(hour//10 == 0 and hour != 9):
            period.append("0" + str(hour) + "-" + "0" + str(hour+1))
        elif(hour//10 == 0 and hour == 9):
            period.append("0" + str(hour) + "-" + str(10))
        else:
            period.append(str(hour) + "-" + str(hour+1))
```

```python
period
```

```python
df['period'] = period
```

```python
df
```

```python
df_crosstab = df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0)
```
